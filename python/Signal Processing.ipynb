```python
def getRotations(d):
    ahrs=MadgwickAHRS(sampleperiod=10)
    q=pd.DataFrame(columns=['yaw','pitch','roll'])
    for idx,row in d.iterrows():
        acc=np.array([row['lastAccelerometerValues[0]'],row['lastAccelerometerValues[1]'],row['lastAccelerometerValues[2]']])
        gyr=np.array([row['lastGyroscopeValues[0]'],row['lastGyroscopeValues[1]'],row['lastGyroscopeValues[2]']])
        mag=np.array([row['lastMagnetometerValues[0]'],row['lastMagnetometerValues[1]'],row['lastMagnetometerValues[2]']])
        ahrs.update(gyr,acc,mag)
        q.loc[idx]=(list(ahrs.quaternion.to_euler_angles()))
    return q
```

```python
def getNormalAcc(d,qx):
    norm_vel=pd.DataFrame(columns=['X','Y','Z'])
    norm_acc=pd.DataFrame(columns=['X','Y','Z'])
    for idx, row in d.iterrows():
        # Find the gravity component of the X-axis

        comp_X = -g * cos(qx.loc[idx].pitch) * sin(qx.loc[idx].roll);

        # Find the gravity component of the Y-axis

        comp_Y = -g * sin(qx.loc[idx].pitch);

        # Find the gravity component of the Z-axis

        comp_Z = g * cos(qx.loc[idx].pitch) * cos(qx.loc[idx].roll);

        norm_acc.loc[idx]=[row['lastAccelerometerValues[0]']-comp_X,row['lastAccelerometerValues[1]']-comp_Y,row['lastAccelerometerValues[2]']-comp_Z]
    return norm_acc
```

```python
def getSequence(d):
    prev='start'
    count=0
    seq=pd.DataFrame(columns=['length','event'])
    for idx,row in d.iterrows():
            curr=row['label'].strip()
            if idx==1:
                continue
            if prev==curr:
                count+=1
            else:

                seq.loc[idx]=[count,prev]
                count=0
            prev=curr
    return seq
```

```python
def getVelPos(norm_acc):
    vel=cumsum(norm_acc.rolling(window=10).sum().fillna(0))
    pos=cumsum(norm_acc.rolling(window=10).sum().rolling(window=10).sum().fillna(0))
    return vel, pos
```

```python
feat=od()
#Feature Set
for key in data.keys():
    d=data[key]
    qx=getRotations(d)
    acc=getNormalAcc(d,qx)
    vel, pos= getVelPos(acc)
    feat[key]=pd.concat([acc[['X','Y']].rename(columns={"X": "LinearAcceleration", "Y": "LateralAcceleration"}),
                      vel[['X','Y']].rename(columns={"X": "LinearVelocity", "Y": "LateralVelocity"}),
                      pos[['X','Y']].rename(columns={"X": "LinearPosition", "Y": "LateralPosition"}),
                      qx[['yaw','roll']].rename(columns={"yaw":"Yaw","roll":"Roll"}),
                      d[['label']]],
                     axis=1,join='inner')
```
