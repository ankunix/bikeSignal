##### 1. Objective

Bike Telemetry Time Series Event Prediction on sequnce data using sensor measurements of gyroscope, accelerometer, magnetometer and GPS position of an Android device mounted on top of a bicycle. The display of these events in form of external visual outputs using LED arrays in order to signal other vehicles about the bike's near-future maneuvers namely turns and stops to fellow riders on other vehicles on road.

#### 2. Experimental Design


- Bicycle on top of which an Android phone is mounted is the primary data collection device, in order for us to interpret the sensor reading into bike maneuvers, the experiment will be conducted in a controlled environment.

- Data is then collected and stored on the Android phone for various trips and binary labels are also stored in form of annotation with corresponding time stamps.

Following satelite image consists of the site where sensor measurements will be done, the white dots represented in the image were used to measure distance on Google Maps.  


Selected parking lot for experimental data collection.
The dimensions measured along the road mid-line are 100 ft x 200 ft.

The selected route is roughly 200 meters long.



#### 3. Data Collection

- Each sample in the time series consists of gyroscope, accelerometer and GPS position, each sequence representing a unique trip.
- Data collection on an Android device's sensors using a reaonable sampling rate. 
- There will be two main classes of events, turn or no turn.
- Each class can be further divided into two sub-classes.

#### 4. Signal Processing

- The collected data is then transferred to a computer with a lot more computational power to apply and test different segmentation, filtering strategies to predict the position and velocity of the bike. This will help identify the relevant features to be utilized for training the sequence classification model.

- The best strategy for sensor fusion will be then deployed on the Android phone to extract features along with the trained classification model to give future predictions.


#### 5. Feature Engineering

Each sample in the time series will consist of 5 attributes :

1. Linear Acceleration $a_X$
2. Lateral Accelaration $a_Y$
3. Roll $g_Z$
4. Velocity $v$
5. GPS Location $l$
6. TimeStamp $t$


- Feature set for representation of each time series consists of first four statistical moments namely __mean__, __standard deviation__, __variance__, and __kurtosis__ for each feature, this reduction in feature space captures only statistically significant features for our learning task.

- Thus each example in the training set will consist of 20 attributes.

#### 6. Learning Task

- The final goal of our model will be to find the most intresting patterns in the feature space to estimate the most likely event.

- Testing on dynamic sample length for 'no turn' events will provide a baseline for selecting a significant fixed sample length for fitting on 'turn' events as well.

- The model selection will be done by 10-fold Cross validation of the ROC-AUC accuracies using 4 classification algorithms, LR(Logistic Regression), NB (Naive Bayes), KNN (K-Nearest Neighbour), and LDA (Linear Discriminant Analysis).

- An ensemble of the best selected models is then used for making the final predictions.

#### 7. Cross Validation

- Dynamic sampling and splitting of sequences across the combination of priors(sequence before an event occurs) and spans(sequence after an event occurs) will provide a holistic view for assessing the model performance across varying sequnce length.
- It will also increase the total number of trained models depending upon variations in the sample length selection.
- Random splitting of the feature space to training(80%) and testing(20%) partitions.
- ROC-AUC is the chosen metric for measuring the cross-validation performance of the ensemble model.

- This will provide a baseline for selecting the best sample length for all events, the problem can then be expanded to include more class labels for turn events left turn & right turn, and for no-turn events in-motion, stop. 

- After satisfactory training iterations and trials, final model will be a multinomial classifier.

#### 8. System Integration
The most accurate trained model representation will then be deployed on the Android which combines the current sensor reading to predict near future events. The model predictions are then sent to a connected RaspberryPi Computer (prefferably wirelessly) which then transmits the visually interprettable bike signals on an LED array, in form of left turn signal, right turn signal, stop signal, no-signal.
